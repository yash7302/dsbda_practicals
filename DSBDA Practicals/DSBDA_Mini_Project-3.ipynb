As there are many NULL values present in the given dataset. We need to replace those values by mean(in case of numerical data) or mode(in case of categorical data)(mode= most frequently occuring value). Here, we need to work on "First Dose Administered" and "Second Dose Administered". Both of them are float, hence we will replace the Nan Values by mean(average).

categorical data ,is data which belongs to specific category or group example : gender , race , country of origin etc
and there are two types of categorical data 1) nominal 2) oridinal 
*nominal data is something which do not naturally have an order or ranking , like gender , race , country of origin etc
*oridinal data is something which have natural ranking or order , example , income level , customer satisfaction rating etc. 

#### For First Dose Administered

```python
# Average of First Dose Administered. finding mean and replacing it with null values 
avg_firstdose = data["First Dose Administered"].astype("float").mean(axis = 0)
print("Average of First Dose:", avg_firstdose)
```

```python
# Replacing First Dose Administered
data["First Dose Administered"].fillna(value = avg_firstdose, inplace=True)
data
```

#### For Second Dose Administered

```python
# Average of Second Dose Administered
avg_seconddose = data["Second Dose Administered"].astype("float").mean(axis = 0)
print("Average of Second Dose:", avg_seconddose)
```

```python
# Replacing Second Dose Administered
data["Second Dose Administered"].fillna(value = avg_seconddose, inplace = True)
data
```

This data is ready to be used for the next questions

### Number of persons state wise vaccinated for first dose in India

```python
first_dose = data.groupby('State')[['First Dose Administered']].sum()
first_dose
```

```python
plt.figure(figsize=(10,13))
plt.barh(data['State'], data['First Dose Administered'] , color="green" )
 
plt.title("Bar Chart")
 
# Setting the X and Y labels
plt.xlabel('State' )
plt.ylabel('First Dose Administered')

# Adding the legends
plt.show()
```
